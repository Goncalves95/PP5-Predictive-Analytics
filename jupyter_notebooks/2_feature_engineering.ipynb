# Import Libraries and Load Data:

Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
from pandas.plotting import scatter_matrix
```

Load Data

```python
# Load the cleaned dataset
df = pd.read_csv('data/dataset.csv')

# Display the first few rows of the dataset
df.head()
```

---

# Univariate analysis:

Distribution of Numerical features

```python
# Plot distribution of numerical features
df.hist(bins=30, figsize=(20, 15))
plt.suptitle('Distribution of Numerical Features')
plt.show()
```

Distribution of Categorical features

```python
# Plot distribution of categorical features
for column in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[column], order=df[column].value_counts().index)
    plt.title(f'Distribution of {column}')
    plt.show()
```

---

# Bivariate Analysis: 

Correlation matrix

```python
# Plot correlation matrix for numeric columns
numeric_df = df.select_dtypes(include=np.number)  # Select numeric columns
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Columns Only)')
plt.show()
```

Scatter plot matrix

```python
# Plot scatter matrix for numerical features
scatter_matrix(df.select_dtypes(include=['float64', 'int64']), figsize=(15, 15))
plt.suptitle('Scatter Matrix for Numerical Features')
plt.show()
```

---

# Multivariate Analysis:

Pairplot

```python
# Plot pairplot for numerical features
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pairplot for Numerical Features')
plt.show()
```

Box pilot 

```python
# Plot box plot for numerical features
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))
plt.title('Box Plot for Numerical Features')
plt.show()
```
